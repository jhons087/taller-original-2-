```python
def impute_nan_meanmedian(df, variable, value):
    if value=="median":
        val=df[variable].median()
    if value=="mean":
        val=df[variable].mean()
    df[variable+"_"+value]=df[variable].fillna(val)
    return df
import random
def rand_float_range(start, end):
    return random.random() * (end - start) + start

def impute_nan_random(df, variable, value=-9):
    if value!=-9:
        df[variable+"_random_fixed"]=df[variable].fillna(value)
    else:
        start,end=df[variable].min(),df[variable].max()
        df[variable+"_random"]=df[variable].fillna(rand_float_range(start,end))
    return df
```

```python
impute_nan_random(data1,"Abs. Win Rate",-9)
```

```python
impute_nan_meanmedian(data1,"Abs. Win Rate","median")
```

```python
impute_nan_meanmedian(data1,"Abs. Win Rate","median")
```

```python
from sklearn.impute import KNNImputer
# define imputer
imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
# fit on the dataset
imputer.fit(data2)
# transform the dataset
Xtrans = imputer.transform(data2)
```

```python
plt.figure(figsize=(10,8))
plt.hist(data1["Abs. Win Rate"], label="original", color="k", histtype= "step", bins=5,lw=8)
plt.hist(data1["Abs. Win Rate_median"],label="imputa median",color="b", histtype= "step", bins=5,lw=4)
plt.hist(data1["Abs. Win Rate_mean"],label="imputa mean",color="r", histtype= "step", bins=5,lw=4)

plt.hist(Xtrans["Abs. Win Rate"],label="KNN", color="y", histtype= "step", bins=5,lw=4)
plt.legend()
plt.show()
```

Visualizando e interpretando los resultados del cuarto periodo de los datos estudiados, podemos decir que la media es la imputación con más bondades.
